```python
#defining a function which will convert all the records of the VisitDate columns to a same format of (YYYY-MM-DD HH:MM:SS) 

def format_date(a):
    Date=[]                                                                  #defining an empty list
    for i in a['VisitDateTime']:                                             #executing for loop to traverse all the columns
        try:                                                                 #try statement 
            c=pd.to_datetime(int(i), unit='ns')                              #converting value to the int datatype first since it's a string then format it using datetime  
            Date.append(c)                                                  # and later appending to the list
        
        except:                                                             # if try statement throws an error, except statement will execute
            Date.append(pd.to_datetime(i,format='%Y-%m-%d %H:%M:%S.%f'))    # it will convert the record to the given format and append to the list 
            
    return Date     #returing the list
```

```python
#calling the format_date function by giving the imput parameter as visitor_data table and adding the new column 'Date' with the appropriate records of datetime


visitor_data['Date']=format_date(visitor_data)
```

```python
# defining a function 'no_of_days_visited_7_days'


def no_of_days_visited_7_days(data):
    data['date_in_datetime']=data.Date.dt.strftime('%d')       #adding a new column "date_in_datetime" in the table, (.dt.strftime('%d')) which will extract the date from the datetime 
    
    #since we need data of only last 7 days we will implement lambda function where Date will be greater than '2018-05-21'
    #then grouping that retrived data on the basis of UserID
    # then finding the unique no of date, because if the user has visited more than 1 time on the same date, we want to include that value as an individual
    days=data[lambda x: x['Date']>='2018-05-21' ].groupby(['UserID']).agg({"date_in_datetime": "nunique"})
    
    #now dropping the 'date_in_datetime' column as we have done with its part
    data=data.drop(['date_in_datetime'],axis=1)
    
    #now merging the 'days' dataframe with the 'data' table on the bais of common 'UserID'
    data=pd.merge(data.sort_values('UserID'),days,on=['UserID'], how='outer')
    
    #renaming 'date_month' column to the respected col name 'No_of_days_Visited_7_Days'
    data=data.rename(columns={'date_in_datetime':'No_of_days_Visited_7_Days'})
    
    #filling the null values of the column by 0 days
    data['No_of_days_Visited_7_Days']=data['No_of_days_Visited_7_Days'].fillna(0)
    return data
```

```python
#now calling the no_of_days_visited_7_days function by inputting the temp table parameter 


temp=no_of_days_visited_7_days(temp)
```

```python
# defining a function 'no_products_viewed_15_days'


def no_products_viewed_15_days(data):
    
    #since we need data of last 15 days we will implement lambda function on the 'data' table
    # now grouping by 'UserID' and agg the data on the unique count of the 'ProductID' and storing it in 'days' dataframe
    days=data[lambda x: (x['Date']>='2018-05-13')].groupby('UserID').agg({"ProductID": "nunique"})
    
    #now storing the 'ProductId' col to the days dataframe
    #NOte: 'ProductID' col in the days dataframe is the count of unique 'ProductID's not the name 
    days=days['ProductID']
    
    # now merging the days dataframe on the basis of 'UserID' and storing it in the 'data' table 
    data=pd.merge(data.sort_values('UserID'),days,on=['UserID'], how='outer')
    
    #Now renaming the given cols to their appropriate name
    data=data.rename(columns={'ProductID_y':'No_Of_Products_Viewed_15_Days','ProductID_x':'ProductID'})
    
    #filling the null values of the column by 0 days
    data['No_Of_Products_Viewed_15_Days']=data['No_Of_Products_Viewed_15_Days'].fillna(0)
    return data    
```

```python
#now calling the no_products_viewed_15_days function by inputting the temp table parameter 


temp=no_products_viewed_15_days(temp)
```

```python
# defining the most_active_os function


def most_active_os(data):
    
    
    #grouping the data on the basis of their 'UserID' and 'OS' and finding the max OS used by the user
    #and storing this in the 'os' dataframe
    os=data.groupby(['UserID'])['OS'].max()
    
    # now merging the os dataframe on the basis of 'UserID' and storing it in the 'data' table 
    data=pd.merge(data.sort_values('UserID'),os,on=['UserID'],how='outer')
    
    #Now renaming the given cols to their appropriate name
    data=data.rename(columns={'OS_y':'Most_Active_OS','OS_x':'OS'})
    return data    
```

```python
#defining the 'pageloads_clicks' function 



def pageloads_clicks(data):
    
    
    #since we need the count of the pageloads and clicks of the last 7 days
    #implement the lambda function which will return the data of last 7 days
    #now grouping that data on the basis of 'UserID' and their corresponding 'Activity' and unstacking the data and their values into new dataframe 'activity'
    activity=data[lambda x: (x['Date']>='2018-05-21')].groupby('UserID')['Activity'].value_counts().unstack()
    
    #now again grouping 'activity' dataframe on the basis of "userID" and 'Pageloads' and storing their sum into 'Pageloads_last_7_days'  
    Pageloads_last_7_days=activity.groupby('UserID')['Pageload'].agg('sum')
    
    #now again grouping 'activity' dataframe on the basis of "userID" and 'Clicks' and storing their sum into 'Clicks_last_7_days'  
    Clicks_last_7_days=activity.groupby('UserID')['Click'].agg('sum')
    
    #now merging that dataframes into the data dataframes
    data=pd.merge(data.sort_values('UserID'), Clicks_last_7_days,on=['UserID'], how='outer')
    data=pd.merge(data.sort_values('UserID'), Pageloads_last_7_days,on=['UserID'], how='outer')
    
    #now renaming the columns to their appropriate names
    data=data.rename(columns={'Click':'Clicks_last_7_days','Pageload':'Pageloads_last_7_days'})
    
    #filling the null values of the column by 0
    data[['Clicks_last_7_days','Pageloads_last_7_days']]=data[['Clicks_last_7_days','Pageloads_last_7_days']].fillna(0,axis=1)       
    return data    
```

```python
##now calling the pageloads_clicks function by inputting the temp table parameter

temp=pageloads_clicks(temp)
```

```python
# Converting '2018-05-28' this to datetime and utc timezone so that datatypes are same
# finding the 'date_diff' of every record by subtracting the 'signupdate' col from the '2018-05-28' date
# now extracting date from 'date_diff' by using timedelta64[D]'
# and storing the data into new col 'user_vintage'

date_diff=pd.to_datetime('2018-05-28',utc='True')-pd.to_datetime(temp['Signup Date'],utc='True')
temp['User_Vintage']=date_diff.astype('timedelta64[D]')
```

```python
def most_viewed_product(data):
    
    # since we need the data of the last 15 days
    # implement lambda function for the specific time period
    # then grouping by the retrieved data on the basis of 'UserID' and 'Activity' and later unstack their values to the new dataframe 'activity_most_viewed'
    activity_most_viewed=data[lambda x: (x['Date']>='2018-05-13')].groupby(['UserID','ProductID'])['Activity'].value_counts().unstack()
    
    #resetting the index
    activity_most_viewed.reset_index(level=1, inplace=True)
    activity_most_viewed.reset_index(level=0, inplace=True)
    
    # then group by the new dataframe on 'UserID' and 'Pageloads' (since we need to deal with only pageloads not click)
    # and aggregating the sum of the data and later storing into the 'Pageloads' dataframe
    Pageloads=activity_most_viewed.drop(['Click'],axis=1)
    
    
    # now merging the 'Pageloads' dataframe to the 'data' table on the 'userID'
    data=pd.merge(data.sort_values('UserID'),Pageloads,on=['UserID'], how='outer')
    
     #now renaming the columns to their appropriate names
    data=data.rename(columns={'ProductID_y':'Most_Viewed_product_15_Days','ProductID_x':'ProductID'})
    
    # now again we will group by 'data' table by 'UserID','Date','Most_Viewed_product_15_Days' and 'Pageload' and store their sum in 'most_viewed_product'
    # we grouped by our data on date because if multiple products have a similar number of pageloads then , we have to consider the recent one
   
    most_viewed_product=data.groupby(['UserID','Date','Most_Viewed_product_15_Days'])['Pageload'].sum()
    
    # now reseting the index of the 'most_viewed_product' as 'pageload'
    most_viewed_product=pd.DataFrame(most_viewed_product.reset_index(name='pageload'))
    
    
    # now again group by data using 'UserID' to find the max 'pageload' and to find their corresponding index in the dataframe     
    most_viewed_product=most_viewed_product.loc[most_viewed_product.groupby('UserID',sort='False')['pageload'].idxmax()]
    
    #dropping the columns which aren't not needed anymore
    data=data.drop(['Most_Viewed_product_15_Days','Pageload'],axis=1)
    
    #dropping the columns
    most_viewed_product=most_viewed_product.drop(['Date','pageload'],axis=1)
    
    #renaming the columns to their relevant names 
    #most_viewed_product=most_viewed_product.rename(columns={'ProductID_y':'Most_Viewed_product_15_Days'})
    
    #now merging the dataframe with the 'data' table on 'userID'
    data=pd.merge(data.sort_values('UserID'),most_viewed_product,on=['UserID'], how='outer')
    
    # now filling the null values by 'Product101'
    data['Most_Viewed_product_15_Days']=data['Most_Viewed_product_15_Days'].fillna('Product101')
    
    #dropping the column
    #data.drop(['Pageload'],axis=1)
    
    return data



temp=most_viewed_product(temp)
    
```

```python
# defining function recently

def recently(data1):
    
    #grouping by the data on the basis of 'UserID' and 'Activity' and later unstack their values to the new dataframe 'recent'
    recent=data1.groupby('UserID')['Activity'].value_counts().unstack()
    
    # then group by the new dataframe on 'UserID' and 'Pageloads' (since we need to deal with only pageloads not click)
    # and aggregating the sum of the data and later storing into the 'Pageloads' dataframe
    Pageloads=recent.groupby('UserID')['Pageload'].agg('sum')
    
    # now merging the 'Pageloads' dataframe to the 'data' table on the 'userID'
    data1=pd.merge(data1.sort_values('UserID'),Pageloads,on=['UserID'], how='outer')
    
    # now again we will group by 'data' table by 'UserID','Date','ProductID' and 'Pageload' and store their sum in 'recent_Viewed_product'
    # we grouped by our data on date because if multiple products have a similar number of pageloads then , we have to consider the recent one
    recent_Viewed_product=data1.groupby(['UserID','Date','ProductID'])['Pageload'].sum()
    
    
    # now reseting the index of the 'recent_Viewed_product' as 'pageload'
    recent_Viewed_product=pd.DataFrame(recent_Viewed_product.reset_index(name='pageload'))
    
    # now retrieving data whose sum of 'pageload' is more than 0 using .loc[] function 
    recent_Viewed_product=recent_Viewed_product.loc[recent_Viewed_product['pageload'].astype(int)>0]
    
    #now group by the 'recent_Viewed_product' dataframe using UserID and then transforming that data on the basis of maximum date i.e. recent one
    #later storing that data into 'recent_Viewed_product' dataframe
    recent_Viewed_product=recent_Viewed_product[recent_Viewed_product.groupby('UserID').Date.transform('max') == recent_Viewed_product['Date']]
    
    
    # now dropping the date column
    recent_Viewed_product=recent_Viewed_product.drop(['Date'],axis=1)
    
    # now merging the dataframe with the 'data' table on 'userID'
    data1=pd.merge(data1.sort_values('UserID'),recent_Viewed_product,on=['UserID'], how='outer')
    
    # now renaming the columns to their appropriate names
    data1=data1.rename(columns={'ProductID_y':'Recently_Viewed_Product','ProductID_x': 'ProductID','Date_x':'Date_y'})
    
    # now filling the null values by 'Product101'
    data1['Recently_Viewed_Product']=data1['Recently_Viewed_Product'].fillna('Product101')
    
    # now dropping the columns which aren't relevant
    data1=data1.drop(['pageload','Pageload'],axis=1)
    return data1

#now calling the function 
temp=recently(temp)
   
```
